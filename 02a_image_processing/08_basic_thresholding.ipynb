# Basic thresholding
Thresholding is a technique of image segmentation. It separates a given single-channel image (or stack) into two regions: Pixels with intensity below a given threshold, also called "background" and pixels with intensity above a given threshold, "foreground".

See also
* [Thresholding (wikipedia)](https://en.wikipedia.org/wiki/Thresholding_(image_processing))
* [Threshold algorithms in scikit-image](https://scikit-image.org/docs/dev/auto_examples/segmentation/plot_thresholding.html)

```python
from skimage.io import imread
import stackview
```

```python
image = imread("data/blobs.tif")
stackview.imshow(image)
```

Thresholding is a basic math operation that works thanks to numpy. It converts an intensity image in a binary image.

```python
binary_image = image > 128

stackview.imshow(binary_image)
```
